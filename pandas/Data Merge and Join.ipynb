# Merge
  - inner
  - outer 
  - left
  - right

# Inner

```python
pd.merge(customers, order)
#iner>>intersection
```

# Outer

```python
pd.merge(customers, order, how = 'outer')
# out>>> union
```

# Left

```python
pd.merge(customers, order, how = 'left')
```

# Right

```python
pd.merge(customers, order, how = 'right')
```

```python
df1 = pd.DataFrame({
    "key": [1,1],
    'left': ['foo', 'bar'],
})

df2 = pd.DataFrame({
    'key': [1,1],
    'right': [7, 8],
})
```

```python
pd.merge(df1, df2, how = 'cross')
```

```python
data1 = {
    "customerID": [1234,1235, 1111,2222,3333],
    'name': ["saad", 'alamgir', 'ali', 'ahmed', 'saadii'],
    'address': ['abc-1', 'xyz-123', 'abc-2', 'abc-3', 'abc-4' ]
}
customers = pd.DataFrame(data1)

data2 = {
    'customer_id': [1234,1235, 1234, 1234,1235, 1267,1237, 1890],
    'order_date': [datetime.date(2023, 5,20), 
                   datetime.date(2023, 5,20),
                   datetime.date(2023, 5,21),
                   datetime.date(2023, 5,20), 
                   datetime.date(2023, 5,22), 
                   datetime.date(2023, 5,20), 
                   datetime.date(2023, 5,23), 
                   datetime.date(2023, 5,20)]
}
order = pd.DataFrame(data2)
```

# Merge by custom collumns

```python
pd.merge(customers, order, left_on='customerID', right_on= 'customer_id')
```

```python
data1 = {
    "customer_id": [1234,1235, 1111,2222,3333],
    'name': ["saad", 'alamgir', 'ali', 'ahmed', 'saadii'],
    'address': ['abc-1', 'xyz-123', 'abc-2', 'abc-3', 'abc-4' ],
    'order_date': [datetime.date(2023, 5,20), 
               datetime.date(2023, 5,20),
               datetime.date(2023, 5,21),
               datetime.date(2023, 5,20), 
               datetime.date(2023, 5,22), 
                  ]
}
customers = pd.DataFrame(data1)

data2 = {
    'customer_id': [1234,1235, 1234, 1234,1235, 1267,1237, 1890],
    'order_date': [datetime.date(2023, 5,20), 
                   datetime.date(2023, 5,20),
                   datetime.date(2023, 5,21),
                   datetime.date(2023, 5,20), 
                   datetime.date(2023, 5,22), 
                   datetime.date(2023, 5,20), 
                   datetime.date(2023, 5,23), 
                   datetime.date(2023, 5,20)]
}
order = pd.DataFrame(data2)
```

```python
data1 = {
    "customer_id": [1234,1235, 1111,2222,3333],
    'name': ["saad", 'alamgir', 'ali', 'ahmed', 'saadii'],
    'address': ['abc-1', 'xyz-123', 'abc-2', 'abc-3', 'abc-4' ],
    'order_date': [datetime.date(2023, 5,20), 
               datetime.date(2023, 5,20),
               datetime.date(2023, 5,21),
               datetime.date(2023, 5,20), 
               datetime.date(2023, 5,22), 
                  ]
}
customers = pd.DataFrame(data1)

data2 = {
    'customer_id': [1234,1235, 1234, 1234,1235, 1267,1237, 1890],
    'order_date': [datetime.date(2023, 5,20), 
                   datetime.date(2023, 5,20),
                   datetime.date(2023, 5,21),
                   datetime.date(2023, 5,20), 
                   datetime.date(2023, 5,22), 
                   datetime.date(2023, 5,20), 
                   datetime.date(2023, 5,23), 
                   datetime.date(2023, 5,20)]
}
order = pd.DataFrame(data2)
```

```python
pd.merge(customers, order, on = "order_date", suffixes=('_left', '_right'))
```
